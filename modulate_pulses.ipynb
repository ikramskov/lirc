```python
import numpy
import ctypes
import cffi
import subprocess
```

```python
subprocess.check_call(['gcc', '-shared', '-fPIC', 'modulate_pulses.c', '-Wall', '-o', 'libmodulate_pulses.so'])

ffi = cffi.FFI()
ffi.cdef("""
typedef uint32_t __u32;

int modulate_pulses(unsigned char* buf, size_t size,
	const int* pulseptr, int n_pulses, __u32 f_sample, __u32 f_carrier, __u32 duty_cycle);
""")
C = ffi.dlopen('./libmodulate_pulses.so')

def modulate_pulses(pulses, f_sample, f_carrier, duty_cycle):
    outbuf = numpy.zeros((65536,), dtype=numpy.ubyte)
    buf = ffi.cast("unsigned char *", outbuf.ctypes.data)
    size = len(outbuf)
    pulseptr = ffi.cast("int *", numpy.array(pulses, dtype=numpy.int32).ctypes.data)
    n_pulses = len(pulses)
    print (buf, size, pulseptr, n_pulses, f_sample, f_carrier)
    outlen = C.modulate_pulses(buf, size, pulseptr, n_pulses, f_sample, f_carrier, duty_cycle)
    print outlen
    return outbuf[:outlen]
```
